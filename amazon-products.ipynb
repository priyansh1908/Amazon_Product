## Cleaning average review rating columns with only ratings and removing the text

```python
dataset['average_review_rating'].unique()
```

```python
def handle_reviews(value):
    if value=='{""text"":""Amazon Cloud Drive""':
        return np.nan
    else:
        value=str(value).split(' ')
        value=value[0]
        return float(value)
        
dataset['average_review_rating']  = dataset['average_review_rating'].apply(handle_reviews) 
dataset.head()
```

## replacing nan values by mean

```python
dataset['average_review_rating'].fillna(dataset['average_review_rating'].mean(),inplace=True)
dataset['average_review_rating'].isnull().sum()
```

```python
dataset['manufacturer'].value_counts()
```

```python
manufac=dataset['manufacturer'].value_counts(ascending=False)
manufac
```

```python
manufacturer_less_than_20=manufac[manufac<20]
manufacturer_less_than_20
```

## rename this as others

```python
def handle_manufacturer(value):
    if (value in manufacturer_less_than_20):
        return 'others'
    else:
        return value
        
dataset['manufacturer']  = dataset['manufacturer'].apply(handle_manufacturer) 
dataset.head()
```

# Working with 'number of reviews' column

```python
dataset['number_of_reviews'].value_counts()
```

```python
dataset['number_of_reviews'].unique()
```

# To remove decimal points so that 1.0 and 1 means the same

```python
def handle_number(value):
    if (value=='subtextKey"":""android-tagline""'):
        value=np.nan
    else:    
        value=str(value).split('.')
        value=value[0]
        return (value)
        
dataset['number_of_reviews']  = dataset['number_of_reviews'].apply(handle_number) 
dataset.head()
```

```python
dataset['number_of_reviews'].value_counts()
```

# Working with 'number of answered questions' column

```python
dataset['number_of_answered_questions'].value_counts()
```

# To remove decimal points so that 1.0 and 1 means the same

```python
def handle_answered(value):
    if value=='panelKey"":""AndroidPanel""}':
        return np.nan
    else:
        value=str(value).split('.')
        value=value[0]
        return float(value)
        
dataset['number_of_answered_questions']  = dataset['number_of_answered_questions'].apply(handle_answered) 
dataset.head()
```

```python
dataset['number_of_answered_questions'].value_counts()
```

# Working with prices columns


```python
dataset['price'].unique()
```

```python
dataset['price'].value_counts()
```

# removing the euro sign to make it numeric

```python
def handle_price(value):
        value=str(value).split()
        value=value[0][1:-1]
        
        #print(values)
        if value=='""subtext"":""' or value=='a' or value=='.':
            value= 0
        if value=='2,439.9':
            value='2439.9'
        return (value)
        
        
        
dataset['price']  = dataset['price'].apply(handle_price) 
dataset['price'] = dataset['price'].astype(float)
#dataset['price'].unique()
dataset.head()
```

# Working with 'category and sub category' column

```python
dataset['amazon_category_and_sub_category'].value_counts()
```

## splitting the category into another column of name category

```python
def handle_category(value):
        value=str(value).split('>')
        value=value[0]
        return (value)
dataset['category']  = dataset['amazon_category_and_sub_category'].apply(handle_category)
dataset.head()
```

## splitting the sub category column

```python
def handle_subcategory(value):
    if value!='nan':
        value=str(value).split('>')
        #print(value)
        #value=value[1]
    else:
        value=value
        
    return (value)
dataset['sub_category']  = dataset['amazon_category_and_sub_category'].apply(handle_subcategory)
dataset.head()
```

```python
category=dataset['category'].value_counts(ascending=False)
category
```

## making a cluster of categories which have value counts less than 10

```python
category_lessthan10=category[category<10]
category_lessthan10
```

```python
def handle_category(value):
    if (value in category_lessthan10):
        return 'others'
    else:
        return value
        
dataset['amazon_category_and_sub_category']  = dataset['amazon_category_and_sub_category'].apply(handle_category) 
dataset.head()
```

# Working with items_customers_buy_after_viewing_this_item column

```python
also_bought=dataset['items_customers_buy_after_viewing_this_item'].value_counts(ascending=False)
also_bought
```

### clustering less than 2 values together as others

```python
also_bought_lessthan2=also_bought[also_bought<2]
also_bought_lessthan2
```

```python

def handle_also_bought(value):
    if (value in also_bought_lessthan2 or value=='NaN'):
        return 'others'
    else:
        return value
        
dataset['items_customers_buy_after_viewing_this_item']  = dataset['items_customers_buy_after_viewing_this_item'].apply(handle_also_bought) 
dataset.head()
```

```python
dataset['number_available_in_stock'].unique()
```

```python
def handle_available(value):
    if value!='text"":""Appstore for Android""' :
        value=str(value).split('\xa0')
        value=value[0]
        value=float(value)
    elif(value=='None'):
        value=0
    else:
        value=0
 
    return float(value)
dataset['number_available_in_stock']  = dataset['number_available_in_stock'].apply(handle_available)
dataset.head()
```
